```python
import glob, pandas as pd, re
```

```python
from sentence_transformers import SentenceTransformer, util
```

```python

model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
paths = glob.glob("data/*_fulltext.csv")
lst = []
for x in paths:
    df = pd.read_csv(x, index_col=0)
    lst +=  [df]
    
df = pd.concat(lst).reset_index(drop = True)
```

```python
df
```

Convert to list of sentences

```python
# df["lst"] = df.fulltext.apply(lambda x : [y.strip() for y in re.split("\.|\!|\?", x)])
```

```python
df["vecs"] = df.fulltext.apply(lambda x : model.encode(x, convert_to_tensor=True))
```

```python
df
```

```python
dct = {}

for i, x in zip(df.fulltext_id, df.vecs):
    for j, y in zip(df.fulltext_id, df.vecs):
        cos = util.cos_sim(x, y)
        if i in dct.keys():
            dct[i].update({j : float(cos)})
        else:
            dct[i] = {j : float(cos)}
```

```python
cosdf = pd.DataFrame(dct)
```

```python
cosdf.to_csv("data/sbert.csv")
```
